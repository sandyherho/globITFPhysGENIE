```python
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
plt.style.use("ggplot")
```

```python
ds_open = xr.open_dataset("../data/open/fields_biogem_2d.nc")
ds_closed = xr.open_dataset("../data/close/fields_biogem_2d.nc")
```

```python
sst = ds_open["ocn_sur_temp"].isel(time=0).to_numpy()
```

```python
lon_open = ds_open["lon_psi"].to_numpy()
lat_open = ds_open["lat_psi"].to_numpy()
psi_open = ds_open["phys_psi"].isel(time=0).to_numpy()
#psi_open[psi_open == 0] = np.nan
```

```python
plt.fill_between(lon_open, lat_open.min(), lat_open.max(), color='#988558');
contour_filled = plt.contourf(lon_open, lat_open, psi_open,
                              #levels=colorbar_levels, 
                              cmap="coolwarm");
contour_lines = plt.contour(lon_open, lat_open, psi_open, 
                            #levels=colorbar_levels, 
                            colors='k', linewidths=0.2);
plt.xlabel(r"Longitude [$^{\circ}$E]");
plt.ylabel(r"Latitude [$^{\circ}$N]");
cbar = plt.colorbar(contour_filled);
```
